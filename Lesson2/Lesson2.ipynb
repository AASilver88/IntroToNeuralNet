```python
import keras
from keras.layers import *
import matplotlib.pyplot as plt
import numpy as np
import keras.utils
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
input_shape = (28, 28, 1)
```

```python
model_1 = keras.Sequential([
      keras.Input(shape=input_shape),
      Flatten(),
      Dense(50),
      Dense(1)
      ] )
```

```python
model_1.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
%%time

history = model_1.fit(x_train, y_train, batch_size=32, epochs=5)
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

```python
model_2 = keras.Sequential([
      keras.Input(shape=input_shape),
      Flatten(),
      Dense(50, activation = 'relu'),
      Dense(10, activation = 'softmax')
      ] )
```

```python
model_2.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
%%time

history = model_2.fit(x_train, y_train_cat, batch_size=32, epochs=50)
```

```python
plt.plot(history.history['accuracy'])
plt.title(' ')
plt.ylabel('')
plt.xlabel(' ')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title(' ')
plt.ylabel('')
plt.xlabel(' ')
plt.show()
```

```python
model_2.evaluate(x_test, y_test_cat)
```

```python
%%time

history = model_2.fit(x_train, y_train_cat, batch_size=32, epochs=5, validation_data =(x_test, y_test_cat) )
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title(' ')
plt.ylabel('')
plt.xlabel(' ')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title(' ')
plt.ylabel('')
plt.xlabel(' ')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
from PIL import Image

def digit(img_path):
    img = Image.open(img_path).convert('L')
    #img = img.resize((28, 28))
    img_array = np.array(img)

    plt.figure(figsize=(2, 2))
    plt.imshow(img_array, cmap='gray')
    plt.show()

    img_array = img_array.reshape(1, 28, 28)
    img_array = img_array / 255
    prediction = model_2.predict(img_array)
    digit = np.argmax(prediction)

    return digit
```
